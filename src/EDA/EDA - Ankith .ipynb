## Maximum size of abnormality. This is a good reference point for GAN

```python
center_x = [x.strip() for x in miasminimeta['center_x'] if type(x) is str]
center_y = [x.strip() for x in miasminimeta['center_y'] if type(x) is str]
```

```python
xC = []
yC = []
for x in center_x:
    try:
        xC.append(int(x))
    except:
        pass
        
for x in center_y:
    try:
        yC.append(int(x))
    except:
        pass

```

```python
%matplotlib inline

plt.scatter(xC,yC)
plt.title('Center of Abnormality in terms of pixel')
```

## Most of the abnormalities are centered around 400-800 pixels in either axis
